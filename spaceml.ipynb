```python
import pandas as pd
df=pd.read_csv("Space_Corrected.csv")
```

```python
df.info()
```

```python
df.drop(["Company Name",'Location','Datum','Detail'],axis=1,inplace=True)
```

```python
df["Status Mission"].value_counts()
```

```python
df.info()
```

```python
df['Status Rocket'].value_counts()
```

```python
def  status(stat):
    if stat=="StatusRetired":
        return 0
    else:
        return 1
df['Status Rocket']=df['Status Rocket'].apply(status)
```

```python
df.info()
```

```python
def cleaning(clean):
    try:
        return float(clean.replace(",",""))
    except:
        return clean
df['Rocket']=df['Rocket'].apply(cleaning)
        
        
```

```python
df.info()
```

```python
a=df['Rocket'].mean()
df['Rocket']=df['Rocket'].fillna(a)
```

```python
df.info()
```

```python
df['Status Mission'].value_counts()
```

```python
df.head()
```

```python
def ozgart(oz):
    if oz=='Success':
        return 5
    elif oz=="Failure":
        return 4
    elif oz=="Partial Failure":
        return 3
    elif oz=='Prelaunch Failure':
        return 2
df['Status Mission']=df['Status Mission'].apply(ozgart)
    
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
x=df.drop('Status Mission',axis=1)
y=df['Status Mission']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(x_train,y_train)
```

```python
y_prediction=classifier.predict(x_test)
print(accuracy_score(y_test,y_prediction)*100)
```

```python
print('Confusion matrix:')
print(confusion_matrix(y_test,y_prediction))
```
